### Introduction

This interactive application guides you through the steps of training and applying WorldCereal classification models.

The app supports three <b>WorldCereal workflows</b>:
- Generate a WorldCereal classification product (map) using pre-trained WorldCereal models for your area and season of interest (only cropland product supported for now).
- Generate a WorldCereal classification product (map) using a custom trained model.
- Train and apply a custom cropland or crop type model and use it to generate a fully customized map for your area, season and crop types of interest.

For training the model, you can use a combination of:
- publicly available reference data harmonized by the WorldCereal consortium;
- your own private reference data.

### Optional: Use a custom presto model

*For advanced users only...*

In case you wish to use a custom Presto model and associated seasonal model zip, you can use the code snippet below to launch the application with your models instead of using the WorldCereal default settings...

````python
from notebook_utils.classification_app import WorldCerealClassificationApp
from pathlib import Path
import urllib.request

from worldcereal.train.backbone import checkpoint_fingerprint

presto_model_path = "url/to/your/presto/..._encoder.pt"
seasonal_model_zip_path = "url/to/your/seasonal_model.zip"

# Download the presto model to your machine to run training and inference
local_pt = Path("./models/local_presto_encoder.pt")
local_pt.parent.mkdir(parents=True, exist_ok=True)
if not local_pt.exists():
    urllib.request.urlretrieve(presto_model_path, local_pt)
# We'll need to explicitly check model's fingerprint here
presto_fp = checkpoint_fingerprint(local_pt)

presto_model_package = {"presto_remote_path": presto_model_path,
                        "presto_local_path": str(local_pt),
                        "presto_fingerprint": presto_fp,
                        "seasonal_model_path": seasonal_model_zip_path}


app = WorldCerealClassificationApp.run(presto_model_package=presto_model_package)
````
