```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

df = pd.read_csv('ESP_LVA_samples_2021_eo_data.csv')
df['valid_time'] = df.valid_date
df['extract'] = 1
df['geometry'] = df.apply(lambda x: Point(x.lon, x.lat), axis=1)

# Pick right columns
df = df[['sample_id', 'ref_id', 'lat', 'lon', 'croptype_name', 'start_date', 'end_date', 'country_code', 'geometry', 'valid_time', 'extract']]

# Make geodataframe
gdf = gpd.GeoDataFrame(df, crs='EPSG:4326')

# Write to geoparquet
gdf.to_file('ESP_LVA_samples_2021_eo_data.gpkg', driver='GPKG')
# gdf.iloc[0:2].to_file('ESP_LVA_samples_2021_eo_data_sample.gpkg', driver='GPKG')
```
