### Introduction

This notebook provides an interactive tool for **querying and exporting WorldCereal extraction datasets** to a single Parquet file that can be consumed directly by the WorldCereal training/finetuning script (`--parquet_files`).

It is the recommended starting point for **[Option 2] Full End-to-End Finetuning** as described in the [finetuning guidelines](../scripts/training/finetuning/finetuning_guidelines.md).

#### What this notebook does

1. **Draw a bounding box** on an interactive map to spatially constrain your query (or leave it empty to query globally).
2. **Choose your data sources:**
   - *Public S3 bucket* — the official WorldCereal public extraction bucket (enabled by default).
   - *Local path(s)* — one or more local Parquet files or directories (e.g. restricted institutional datasets or custom extraction outputs). You can add as many paths as needed.
3. **Filter options** — optionally restrict the query to temporary crop samples only (default: all land-cover categories).
4. **Inspect the results** — a summary table shows the number of samples and crop types per dataset.
5. **Save** the merged, deduplicated result to a single Parquet file of your choice.

<div class="alert alert-block alert-info">
<b>Note on deduplication:</b><br>
When combining the public S3 bucket with local paths, samples that exist in both sources are automatically deduplicated based on <code>sample_id</code> and <code>timestamp</code>, ensuring that no observation is counted twice.
</div>
